## CSV2df

This function will allow us to standardise the way we are importing our data. na_values will treat the values in the list as missing values, I set the data type to be none, and specifies that the character encoding to be latin-1 before returning the variable df which stores the entire dataset values.

```python
# CSV2df

def CSV2df(filename):
    df = pd.read_csv('data/' + filename, na_values=['NA', 'N/A'], dtype=None, encoding='latin-1')
    return df
```

## simpleInfo

This function 

```python
# simpleInfo

def simpleInfo(df, filename, info):
    print(f'Name of file: {filename}' + '.csv')
    print(info)
    print(f'This file have {df.shape[0]} rows and {df.shape[1]} columns\n')
    print(f'Number of missing values in each columns:\n{df.isnull().sum()}\n')
    print(f'Statistics on columns with interger:\n{df.describe()}\n')
    print(f'First few row of data:\n{df.head}\n')
```

## Graph 1 - Bar x Line Graph

To start things off, Bob wants to find out the relationship between the number of breaches and affected individuals over the years. He wants to see if there are any trend on both variables and determine if there is a pattern.

```python
# Graph 1 - Code

# data extraction for line plot 
# extracting the dates and only giving me the years (x axis)
unique_years = data_breaches_all_clean['Breach Submission Date'].str.split('/').str[2].unique() 
# reversing unqiue years so that it starts from 2009 instead of 2021
unique_years = unique_years[::-1]

# extracting the number of affected individual by year
individuals_affected = data_breaches_all_clean.groupby(data_breaches_all_clean['Breach Submission Date'].str.split('/').str[2]).sum()['Individuals Affected']

# data extraction for bar graph
# extracting number of breach by years
num_of_breaches = data_breaches_all_clean.groupby(data_breaches_all_clean['Breach Submission Date'].str.split('/').str[2]).size()


# Plotting da graph :D 

# plotting both on the same graph with different y-axes
fig, ax1 = plt.subplots(figsize=(12, 8))

# plotting line plot on the left y-axis

ax1.plot(unique_years, individuals_affected, marker='o', label='Affected Individuals (in thousands)', color='orange')
ax1.set_xlabel('Year')
ax1.set_ylabel('Sum of Affected Individuals (in thousands)', color='orange')
ax1.tick_params(axis='y', labelcolor='orange')
# format y-axis tick labels to display in thousands
ax1.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,.0f}".format(x)))
for i, v in enumerate(individuals_affected):
    ax1.text(unique_years[i], v, "{:,.0f}".format(v), color='black', ha='center', va='bottom', rotation=20)

# creating a second y-axis on the right for the bar graph
ax2 = ax1.twinx()
ax2.bar(unique_years, num_of_breaches, color='skyblue', label='Number of Breaches', alpha=0.7)
ax2.set_ylabel('Number of Breaches', color='skyblue')
ax2.tick_params(axis='y', labelcolor='skyblue')
for i, v in enumerate(num_of_breaches):
    ax2.text(unique_years[i], v + 5, str(v), color='blue', ha='center', va='bottom')

ax2.legend(loc='upper right')

# combining the legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')
plt.title('Sum of Affected Individuals and Number of Breaches by Year')
plt.show()
```

## Graph 2 - Pie Chart

Now that Bob know that the number of data breaches is increasing over the years, he wonder what type of businesses are being targetted more as well as the frequency of the common type of breach that is affecting each type of businesses.

```python
# Graph 2 - Code

# data extraction
# extracting the type of business (unqiue)
covered_entity_clean, covered_entity_breaches = data_breaches_all_clean[['Covered Entity Type']], breach_report[['Covered Entity Type']]
unique_entity = pd.concat([covered_entity_clean, covered_entity_breaches], ignore_index=True)['Covered Entity Type'].dropna().unique()
unique_entity = unique_entity[:-1] # removed the last element as it have little data

# extracting type of breaches
# extracting unique types of breach
unique_breaches_report = breach_report['Type of Breach'].unique()
unique_breaches_clean = data_breaches_all_clean['Type of Breach'].unique()
split_breaches_report = [x.split(',')[0].strip() if isinstance(x, str) else x for x in unique_breaches_report]
split_breaches_clean = [x.split(',')[0].strip() if isinstance(x, str) else x for x in unique_breaches_clean]
# combining and getting unique elements
unique_breaches = list(set(split_breaches_report + split_breaches_clean))

# i will only be taking the major type of breaches

chosen_breach_type = ['Hacking/IT Incident', 'Improper Disposal', 'Theft', 'Unauthorized Access/Disclosure']

combined_data = pd.concat([data_breaches_all_clean[['Type of Breach', 'Covered Entity Type']], breach_report[['Type of Breach', 'Covered Entity Type']]], axis=0, ignore_index=True)
combined_data['Count'] = combined_data.groupby(['Type of Breach', 'Covered Entity Type'])['Type of Breach'].transform('count')
combined_data = combined_data.drop_duplicates().dropna()

# extracting count of each chosen breach type in each business
filtered_breach_counts = combined_data[
    (combined_data['Type of Breach'].isin(chosen_breach_type)) &
    (combined_data['Covered Entity Type'].isin(unique_entity))
]


# preparing to plot the inner layer of the pie chart
reshaped_filtered = filtered_breach_counts.pivot(index='Covered Entity Type', columns='Type of Breach', values='Count').fillna(0).values.tolist()

# plotting da graph

fig, ax = plt.subplots(figsize=(6.5, 6.5))
size = 0.3
outer_colors = plt.colormaps["tab20b"]([5, 9, 13])
inner_colors = plt.colormaps.get_cmap('tab20c')([1, 5, 9, 13])

outer_pie = ax.pie(np.sum(reshaped_filtered, axis=1), radius=1, colors=outer_colors,
       wedgeprops=dict(width=size, edgecolor='w'), labels=['Business Associate', 'Health Plan', 'Healthcare Provider'], startangle=90, autopct='%1.1f%%', pctdistance=0.85, explode=[0, 0, 0.05])

inner_pie = ax.pie(np.array(reshaped_filtered).flatten(), radius=1-size, colors=inner_colors,
       wedgeprops=dict(width=size, edgecolor='w'), startangle=90, autopct='%1.1f%%', pctdistance=0.85, textprops={'fontsize': 6.5}, explode=[0, 0, 0, 0, 0, 0, 0, 0, 0.05, 0.05, 0.05, 0.05])

ax.legend(inner_pie[0], ['Hacking Incident', 'Improper Disposal', 'Theft', 'Unauthorized Access'], title="Breach Types", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

ax.set(aspect="equal", title='Percentage of breach type')
plt.show()
```

### Terminology

After that, I used the dropna and unique function to drop all rows with empty the column Covered Entity Type empty and take the unique value, which I ended up with 4 values. Since the last element 'Healthcare Clearing House' have very little data, I decided to drop it.

Moving on, I extracted the type of breaches and used the unique function to get the unique values. Hacking/IT Incident represent vulnerable systems, improper disposal represent employees lacking of knowlegde, theft representing vulnerable physical security, and unauthorised access/disclosure represent insider.

## Graph 4 - Pie chart

Now that Bob know that the number of CVE is constantly on a upward trend and predicts that it will not go down much, he wants to find out which attack vector is always exploited during a data breach.

```python
# Graph 4 - Code

# extracting the attack vector that are involved in hacking
# group by 'Type of Breach' and 'Location of Breached Information' then count occurrences
# for breach_report
hacking_data_report= breach_report[breach_report['Type of Breach'] == 'Hacking/IT Incident']
grouped_data_report = hacking_data_report.groupby(['Type of Breach', 'Location of Breached Information']).size().reset_index(name='Count')
# for data_breaches_all_clean
hacking_data_clean = data_breaches_all_clean[data_breaches_all_clean['Type of Breach'] == 'Hacking/IT Incident']
grouped_data_clean = hacking_data_clean.groupby(['Type of Breach', 'Location of Breached Information']).size().reset_index(name='Count')

attack_vector = pd.concat([grouped_data_clean, grouped_data_report], ignore_index=True)
# split by ',' and take the first element
attack_vector['Location of Breached Information'] = attack_vector['Location of Breached Information'].apply(lambda x: x.split(',')[0].strip())

# group by 'Type of Breach' and 'Location of Breached Information', then sum the 'Count' column
attack_vector = attack_vector.groupby(['Type of Breach', 'Location of Breached Information'])['Count'].sum().reset_index()

# plotting da graph
colors = plt.cm.Set2(range(len(attack_vector)))

plt.figure(figsize=(8, 8))
pie = plt.pie(attack_vector['Count'], autopct='%1.1f%%', startangle=90, colors=colors, pctdistance=1.1, textprops={'rotation': 25})

plt.title('Distribution of Attack Vectors for Hacking/IT Incident')
plt.legend(pie[0], attack_vector['Location of Breached Information'], loc="center left", bbox_to_anchor=(1, 0.8))
plt.show()
```

### Terminology

For the extraction of data, I had to filter out to ensure that only those hacking incident are in my dateframe. After that, I count the number of occurance before splitting those locations with commas. I combined similar names together and add the counts so that I can get the number of types a hacking incident occur with that location.

For the plotting, since I am just doing a simple pie chart, it was relatively easy. The only issue I face was that the others and paper/films section does not take up much percentage and are beside one another, making the names overlapped. To fixed that, I slanted the labels.

## Graph 5 - Bar graph

Since network server appear at the top of the attack vector list, Bob decided to find out more about the vulnerabilities that each CVE is exploiting.
```python
# Graph 5 - Code

cve_name = cve[cve['summary'].str.contains('\*\* REJECT \*\*') == False]
cve_name = cve_name['cwe_name']
vulnerabilities = ['Buffer Overflow', 'Web', 'Configuration Issue', 'DDOS', 'Cryptographic Failure', 'Insecure Design']
vuln_dict = {key: [] for key in vulnerabilities}
unincluded = []

# 'ai' part to sort the differnet vulnerability according to a broader term
for indexes, name in enumerate(cve_name):
    included = False

    for vuln in vulnerabilities:
        if vuln.lower() in name.lower():
            vuln_dict[vuln].append(name)
            included = True
            break
        if not included:
            if 'flow' in name.lower() or 'buffer' in name.lower():
                vuln_dict['Buffer Overflow'].append(name)
                included = True
            elif 'CSRF' in name or 'cross-site scripting' in name.lower() or 'injection' in name.lower() or 'xml' in name.lower() or 'path' in name.lower() or 'url' in name.lower():
                vuln_dict['Web'].append(name)
                included = True
            elif 'permission' in name.lower() or 'check' in name.lower() or 'SSRF' in name or 'config' in name.lower() or 'unrestrict' in name.lower():
                vuln_dict['Configuration Issue'].append(name)
                included = True
            elif 'resource' in name.lower() or 'excess' in name.lower() or 'loop' in name.lower() or 'limit' in name.lower():
                vuln_dict['DDOS'].append(name)
                included = True
            elif 'crytography' in name.lower() or 'password' in name.lower() or 'algorithm' in name.lower() or 'cryptographic' in name.lower() or 'creden' in name.lower() or 'encrypt' in name.lower() or 'info' in name.lower() or 'valid' in name.lower():
                vuln_dict['Cryptographic Failure'].append(name)
                included = True
            elif 'exposure' in name.lower() or 'authen' in name.lower() or 'improper' in name.lower() or 'authori' in name.lower() or 'incorrect' in name.lower() or 'incomplete' in name.lower() or 'error' in name.lower() or 'condition' in name.lower() or 'function' in name.lower() or 'force' in name.lower() or 'code' in name.lower():
                vuln_dict['Insecure Design'].append(name)
                included = True
        
        if not included:
            unincluded.append(name)


# getting ready to plott
num_of_vuln_type = []
vuln_type_name = []
for key, value in vuln_dict.items():
    count = len(value)
    num_of_vuln_type.append(count)
    vuln_type_name.append(key)

# plotting da graph

bars = plt.bar(vuln_type_name, num_of_vuln_type, color='skyblue')
plt.xlabel('Vulnerability type')
plt.ylabel('Count')
plt.xticks(rotation=20, ha='right')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom')
plt.title('Number of vulnerability type from 2007 to 2019')
plt.show()
```

After that, I had to come up with a few common category of hacking and did an 'ai' part, where I sort all 85,000 ++ data according to their category in a dictionary. I left out a few as they can appear on multiple category or they can be a category on their own.

After that, all I had to do was to take the dictionary key and value to form a list, and parse it into the matplotlib function to form a graph.

```python
# Graph 6 - Code

# data extraction
data = cve_2021[['Confidentiality_impact', 'Integrity_impact', 'Availability_Impact', 'Base Score', 'Exploitability_Score', 'Impact_score']].dropna()


# plotting da graph

def determine_size(impact_score):
    if impact_score < 1:
        return 10
    elif impact_score < 2:
        return 40
    elif impact_score < 3:
        return 70
    elif impact_score < 4:
        return 100
    elif impact_score < 5:
        return 130
    else:
        return 160
    
def determine_color(impact):
    c = impact["Confidentiality_impact"]
    i = impact['Integrity_impact']
    a = impact['Availability_Impact']

    if c == 'HIGH' and i != 'HIGH' and a != 'HIGH': # only c
        return '#e377c2' # Pink
    elif c != 'HIGH' and i == 'HIGH' and a != 'HIGH': # only i
        return '#8c564b' # Brown
    elif c != 'HIGH' and i != 'HIGH' and a == 'HIGH': # only a
        return '#bcbd22' # Yellow
    elif c == 'HIGH' and i == 'HIGH' and a != 'HIGH': # only c and i
        return '#ff7f0e' # Orange
    elif c == 'HIGH' and i != 'HIGH' and a == 'HIGH': # only c and a
        return '#1f77b4' # Blue
    elif c != 'HIGH' and i == 'HIGH' and a == 'HIGH': # only i and a
        return '#9467bd' # Purple
    elif c != 'HIGH' and i != 'HIGH' and a != 'HIGH': # none
        return '#2ca02c' # Green
    else: return '#d62728' # all # Red

# create new columns
data['sizes'] = data['Impact_score'].apply(determine_size)
data['color'] = data.apply(determine_color, axis=1)
data = data.dropna()
plt.figure(figsize=(11, 5))
colors_array = data['color'].values
plt.scatter(data['Base Score'], data['Exploitability_Score'], c=colors_array, s=data['sizes'], alpha=0.1)
plt.title('2021 CVE Analysis', pad=30)
plt.xlabel('Base Score')
plt.ylabel('Exploitability Score')

# make custom legends base on CIA and color
legends = [
    {'marker': 's', 'color': 'red', 'label': 'Highly Dangerous'}, 
    {'marker': 's', 'color': 'orange', 'label': 'Dangerous'}, 
    {'marker': 's', 'color': 'blue', 'label': 'Dangerous'}, 
    {'marker': 's', 'color': 'purple', 'label': 'Dangerous'}, 
    {'marker': 's', 'color': 'pink', 'label': 'Moderately Dangerous'},
    {'marker': 's', 'color': 'brown', 'label': 'Moderately Dangerous'},
    {'marker': 's', 'color': 'yellow', 'label': 'Moderately Dangerous'},
    {'marker': 's', 'color': 'green', 'label': 'Not Dangerous'}
]

# first legend for colors
handles_colors = [Line2D([0], [0], marker=d['marker'], color=d['color'], label=d['label'], markersize=6) for d in legends]
legend_colors = plt.legend(handles=handles_colors, loc='upper left')

# second legend for sizes
start_size = 5
sizes_legend = [Line2D([0], [0], marker='o', color='black', markersize=start_size + i, label=f'Impact {i + 0} - {i + 1}') for i in range(6)]
legend_sizes = plt.legend(handles=sizes_legend, loc='upper center', fontsize='small', ncol=len(sizes_legend), bbox_to_anchor=(0.5, 1.1))

# combine legends
plt.gca().add_artist(legend_colors)

plt.grid(True)
plt.show()
```
